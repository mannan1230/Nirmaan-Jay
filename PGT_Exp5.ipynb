```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator, BayesianEstimator
```

```python
heart.columns
```

```python
Model = BayesianNetwork([('age','trestbps'),('age','fbs'),('sex','trestbps'),('exang','trestbps'),('exang','trestbps'),('trestbps','target'),('target','restecg'),('fbs','target'),('target','thalach'),('target','chol')])
```

```python
Model.fit(heart, estimator=MaximumLikelihoodEstimator)
```
